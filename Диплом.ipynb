```python
# Initial imports
import pydot
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.tree import export_graphviz
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
# Read the file


df  = cleaned_df
df .head()
```

```python
# Rename columns
df.columns = ["Date", "Total_Production", "Product_Supplied", "Refinery_Input", "Operable_Dist_Capacity", "Operating_Dist_Capacity", "Idle_Dist_Capacity", "Percent_Util", "Future_Price", "Spot_Price"]
df.head()
```

```python
# Get month and Year from Date
df["Year"] = pd.DatetimeIndex(df["Date"]).year
df["Month"] = pd.DatetimeIndex(df["Date"]).month

# Reorder Columns
df = df[["Date", "Year", "Month", "Total_Production", "Product_Supplied", "Refinery_Input", "Operable_Dist_Capacity", "Operating_Dist_Capacity", "Idle_Dist_Capacity", "Percent_Util", "Future_Price", "Spot_Price"]]

# Sorting DataFrame by Year and Month
df.sort_values(["Year", "Month"], ascending=[True, True], inplace=True)

# DataFrame without index
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
print("The shape of our DataFrame is:", df.shape)
```

```python
# Descriptive statistics for each column
df.describe()
```

## Предварительная обработка данных

```python
# Define the features and target set.
X = df.drop(["Date", "Spot_Price", "Future_Price"], axis = 1)
y = np.array(df["Spot_Price"])

# Saving feature names for later use
X_list = list(X.columns)

# Convert to numpy array
X = np.array(X)
```

## Наборы для обучения и тестирования

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
print("Training Features Shape:", X_train.shape)
print("Training Labels Shape:", y_train.shape)
print("Testing Features Shape:", X_test.shape)
print("Testing Labels Shape:", y_test.shape)
```

```python
# Создаем экземпляр StandardScaler.
scaler = StandardScaler()
# # Подгоняем стандартную шкалу к обучающим данным.
X_scaler = scaler.fit(X_train)

# Масштабируем данные.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Соответствует модели случайного леса

```python
# Создаем случайный лесной регрессор
rf_model = RandomForestRegressor(n_estimators = 1000, random_state = 42)

```

```python
rf_model
```

```python
# Подгонка модели
rf_model = rf_model.fit(X_train_scaled, y_train)
```

## прогнозы на основе тестового набора

```python
# Составление прогнозов с использованием данных тестирования
y_pred = rf_model.predict(X_test_scaled)
```

## Оценка модели

```python
# Вычислите среднее абсолютное значение (MAE),
# среднеквадратичное значение (MSE) и среднеквадратичную ошибку (RMSE).

print(f"Mean Absolute Error(MAE): {metrics.mean_absolute_error(y_test, y_pred):.2f}")
```

```python
# Абсолютные погрешности (MAE)
errors = abs(y_pred - y_test)
# Вычислить среднюю абсолютную процентную ошибку (MAPE)
mape = 100 * (errors / y_test)
# Точность расчета и отображения
accuracy = 100 - np.mean(mape)
print("Accuracy:", round(accuracy, 2), "%.")
```

## Визуализация единого дерева решений

```python
# Вырвите одно дерево из леса.
tree = rf_model.estimators_[5]

# Экспортируйте изображение в точечный файл
export_graphviz(tree, out_file = "tree.dot", feature_names = X_list, rounded = True, precision = 1)
# Используйте точечный файл для создания графика
(graph, ) = pydot.graph_from_dot_file("tree.dot")
# Записать график в png-файл
graph.write_png("tree.png")
```

## Дерево уменьшенного размера

(Чтобы упростить задачу: уменьшите глубину расположения деревьев в лесу для получения понятного изображения)

```python
# Augmented Dickey-Fuller Test. Наши данные не кажутся стационарными, но мы должны это подтвердить.
def adf_test(dataset):
  dftest3 = adfuller(dataset, autolag = 'AIC')
  print("1. ADF : ",dftest3[0])
  print("2. P-Value : ", dftest3[1])
  print("3. Количество лагов : ", dftest3[2])
  print("4. Количество наблюдений, использованных для регрессии ADF и расчета критических значений :", dftest3[3])
  print("5. Критические значения :")#Критические значения: Критические значения для теста Дики-Фуллера при различных уровнях значимости (1%, 5% и 10%).
          #Эти значения используются для сравнения с вычисленной статистикой теста (ADF) для принятия решения о стационарности ряда
  for key, val in dftest3[4].items():
      print("\t",key, ": ", val)

#ADF: Это расширенный тест Дики-Фуллера (Augmented Dickey-Fuller test),
#который показывает, является ли временной ряд стационарным или нет. Значение статистики теста Дики-Фуллера.
# P-Value: Значение вероятности (p-value) для теста Дики-Фуллера.
# Если p-value меньше выбранного уровня значимости (обычно 0.05),
#то нулевая гипотеза о нестационарности временного ряда отвергается, и ряд считается стационарным.
#Количество лагов, использованных в расширенном тесте Дики-Фуллера. Это определяется автоматически при использовании параметра autolag='AIC'.
```

## Batch Size 4, window length 12

```python
#определение генераторов обучения и тестирования
window_length=12
batch_size=2
num_features=6
training_generator = TimeseriesGenerator(x_train, y_train, length=window_length, sampling_rate=1, batch_size=batch_size)
testing_generator = TimeseriesGenerator(x_test, y_test, length=window_length, sampling_rate=1, batch_size=batch_size)
```
